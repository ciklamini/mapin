```python
# import gmplot package
import gmplot

latitude_list = [48.9777042, 48.9779928, 48.9781989]
longitude_list = [14.4689864, 14.4700217, 14.4687556]
gmap3 = gmplot.GoogleMapPlotter(48.9777042,
                                14.4689864, 13)

# scatter method of map object 
# scatter points on the google map
gmap3.scatter( latitude_list, longitude_list, '# FF0000',
                            #   size = 40, 
                              size = [30,40, 50], 
                              marker = False )

# Plot method Draw a line in
# between given coordinates
gmap3.plot(latitude_list, longitude_list, 
           'cornflowerblue', edge_width = 2.5)

gmap3.draw( "map13.html" )
```

```python
node_position_str = {"node 1": "48.9777042N, 14.4689864E",
"node 2": "48.9779928N, 14.4700217E",
"node 3": "48.9781989N, 14.4687556E"}
```

```python
def parse_node_positions(node_position_str):
    latitude_list = []
    longitude_list = []
    for pos in node_position_str.values():
        lat_str, lon_str = pos.split(',')
        lat = float(lat_str[:-1])
        if lat_str[-1] == 'S':
            lat = -lat
        lon = float(lon_str[:-1])
        if lon_str[-1] == 'W':
            lon = -lon
        latitude_list.append(lat)
        longitude_list.append(lon)
    return latitude_list, longitude_list

# Example usage:
latitudes, longitudes = parse_node_positions(node_position_str)
print(latitudes)
print(longitudes)
```

```python
gmap3.heatmap( latitude_list, longitude_list )
gmap3.draw( "map14.html" )
```

```python
import dash
from dash import dcc, html

import plotly.graph_objects as go

# Create a Plotly figure with the existing coordinates
fig = go.Figure()

# Add scatter points
# fig.add_trace(go.Scattermapbox(
fig.add_trace(go.scattermap(
    lat=latitudes,
    lon=longitudes,
    mode='markers+lines',
    marker=dict(size=15, color='red'),
    line=dict(width=3, color='cornflowerblue'),
    text=list(node_position_str.keys()),
    hovertemplate='<b>%{text}</b><br>Lat: %{lat}<br>Lon: %{lon}<extra></extra>'
))

# Update layout for mapbox
fig.update_layout(
    mapbox=dict(
        style="open-street-map",
        center=dict(lat=latitudes[0], lon=longitudes[0]),
        zoom=13
    ),
    height=600,
    margin=dict(l=0, r=0, t=0, b=0)
)

# Create Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Interactive Map Visualization"),
    dcc.Graph(
        id='map',
        figure=fig
    )
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
